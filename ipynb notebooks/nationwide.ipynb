```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import requests
import time
import scipy.stats as st
from scipy.stats import linregress
```

```python
# Load the CSV file created in Part 1 into a Pandas DataFrame
midwest_df = pd.read_csv("../data/midwest_data.csv")
southeast_df = pd.read_csv("../data/southeast_data.csv")
southwest_df = pd.read_csv("../data/southwest_data.csv")
northeast_df = pd.read_csv("../data/northeast_data.csv")
west_df = pd.read_csv("../data/west_data.csv")

# Display sample data
midwest_df["Region"] = "Midwest"
southeast_df["Region"] = "Southeast"
southwest_df["Region"] = "Southwest"
northeast_df["Region"] = "Northeast"
west_df["Region"] = "West"
west_df.head()
```

```python
# Merge two DataFrames
census_pd = pd.concat([midwest_df, southeast_df, southwest_df, northeast_df, west_df])
census_pd
```

```python
# Save the DataFrame as a CSV
# Note: To avoid any issues later, use encoding="utf-8"
census_pd.to_csv("../data/nationwide_data.csv", encoding="utf-8", index=False)
```

```python
data_years = [
    2018,
    2019,
    2020,
    2021,
    2022
]
```

```python
# Add the columns to get the total number of owned homes
census_pd["Total Owned Homes"] = census_pd["Total Housing with Mortgage"] + census_pd["Total Housing without Mortgage"]

# Calculate the home ownership rate
census_pd["Home Ownership Rate"] = 100 * census_pd["Total Owned Homes"].astype(int) / census_pd["Population"].astype(int)

# Group by year and calculate the mean home ownership rate for each year, and rename it
yearly_ownership_rate = census_pd.groupby("Year")["Home Ownership Rate"].mean()
yearly_ownership_rate.head()
```

```python
def regressValues(x_values, y_values):
    (slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
    regress_values = x_values * slope + intercept
    return regress_values
```

```python
# Create boxplot for Poverty Rate by Year
plt.figure(figsize=(12, 6))
census_pd.boxplot(column="Poverty Rate", by="Year")

# Customize the plot
plt.title('Poverty Rate Distribution by Year')
plt.xlabel('Year')
plt.ylabel('Poverty Rate (%)')
plt.grid(False)
#plt.suptitle('')  # Suppress the default title to keep it clean

# Show the plot
plt.show()

# Calculate and list the outliers with states
outliers_list = []

for year in census_pd["Year"].unique():
    year_data = census_pd[census_pd["Year"] == year][["State", "Poverty Rate"]]
    Q1 = year_data["Poverty Rate"].quantile(0.25)
    Q3 = year_data["Poverty Rate"].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = year_data[(year_data["Poverty Rate"] < lower_bound) | (year_data["Poverty Rate"] > upper_bound)]
    
    if not outliers.empty:
        outliers_list.append((year, outliers))

# Print the outliers
print("Outliers by Year and State")
for year, outliers in outliers_list:
    print(f"\nYear {year}:")
    for index, row in outliers.iterrows():
        print(f"  State: {row['State']}, Poverty Rate: {row['Poverty Rate']}")
```
